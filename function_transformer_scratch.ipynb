```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
X=np.array([[1,2],[3,4]])
log_transform=FunctionTransformer(np.log1p)
X_transformed=log_transform.transform(X)
print(X_transformed)
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
X=np.array([[1,2],[3,4]])
def squ(X):
  return np.hstack((X,X**2))
custom_transform=FunctionTransformer(squ)
X_transformed=custom_transform.transform(X)
print(X_transformed)
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
X=np.array([[1,2],[3,4]])
def my_scaling(X):
  return X/np.max(X)
custom_transform=FunctionTransformer(my_scaling)
X_transformed=custom_transform.transform(X)
print(X_transformed)
```

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np
X=np.array([[1,2],[3,np.nan]])
def my_cleaing(X):
  X[np.isnan(X)]=0
  return X
custom_transform=FunctionTransformer(my_cleaing)
X_transformed=custom_transform.transform(X)
print(X_transformed)
```
