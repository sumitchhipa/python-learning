```python
#pos tagging to nlp
# pos tagging means part of speech tagging
# it is an nlp techinique where each word is a sentance is assigned its grametical role such as
# noun
# verb
# adjective
# adverb
# pronoun


# why pos tagging is important
# pos tagging is used in
# text calssification
# chatbots
# machine Transformation
# named entity recognition
# sentiment analysis
# speech recogntion
# grammmer checking
```

```python
import spacy
# load english model
nlp = spacy.load('en_core_web_sm')
# input text
doc = nlp("sumit is learning NLP")
# pos tagging
for token in doc:
  print(token.text,"--->",token.pos_)
```

```python
#example --
import spacy
nlp = spacy.load("en_core_web_sm")
text = nlp("sumit is learning NLP")
doc = nlp(text)
for token in doc:
  print(
      "word",token.text,
      "| POS",token.pos_,
      "| TAG",token.tag_,
      "| DEP",token.dep_,
      "| LEMMA",token.lemma_
  )
```

```python
# visualizing POS TAgging
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
text = nlp("sumit is learning NLP")
doc = nlp(text)
displacy.serve(doc,style="dep")
```
