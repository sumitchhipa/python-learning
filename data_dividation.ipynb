```python
# data dividation
# df---> x+y
# x---> x_train ,x_test
# y---> y_train ,y_test
import numpy as np
import pandas as pd
df=pd.read_csv("/content/placement.csv")
df.head()
```

```python
x=df.drop(columns=["placed"])#in[putdata]
y=df["placed"]#target data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
np.round(x_train.describe(),2)
x_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
np.round(x_train_sc)
type(x_train_sc)
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
type(x_train_new)
x_train_new.head()
```

```python
df=pd.read_csv("/content/covid_toy.csv")
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer()
df['fever']=si.fit_transform(df[['fever']])
df.isnull().sum()
```

```python
df['has_covid'].value_counts()
```

```python
df['gender']=df['gender'].map({'Male':1,'Female':0})
df['cough']=df['cough'].map({'Mild':1,'Strong':2,'Strong':3})
df['city']=df['city'].map({'Delhi':1,'Mumbai':2,'Bangalore':3})
df['has_covid']=df['has_covid'].map({'Positive':1,'Negative':0})
df.head()
```

```python
x=df.drop(columns=['has_covid'])#impute data
y=df['has_covid']#target data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```
