```python
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Bidirectional, Dense, Embedding
from tensorflow.keras.datasets import imdb
import numpy as np
from tensorflow.keras.models import Sequential
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=5000)
# getting review with words that come under 5000
# most occouring words in the entire
# corpus of textual review data
```

```python
print(x_train[0])
```

```python
# getting all the words from word index dictionary
word_idx = imdb.get_word_index()
# orignally the index number of a value and not a key,
# hence converting the index as a key and the words as value
word_idx ={i: word for word, i in word_idx.items()}
# again printing the review
print([word_idx[i] for i in x_train[0]])
```

```python
# Get the minimum and maximum length of review
print("Max length of a review", len(max((x_train + x_test), key=len)))
print("Min length of a review", len(min((x_train + x_test), key=len)))
```

```python
from tensorflow.keras.preprocessing import sequence
# keeping a  fixed lentgth of all reviews to max words
max_words = 400
x_train = sequence.pad_sequences(x_train, maxlen=max_words)
x_test = sequence.pad_sequences(x_test, maxlen=max_words)
x_valid, y_valid = x_train[:64],y_train[:64]
x_train, y_train = x_train[64:],y_train[64:]
```

```python
# fixing every word's embedding size to 32
embed_len = 32
# Define vocab_size based on num_words used in imdb.load_data
vocab_size = 5000
# creating a RNN model
RNN_model = Sequential(name='Simple_RNN')
RNN_model.add(Embedding(vocab_size,
                        embed_len))
# in case of a stacked (more than one layer of RNN)
# use return_sequences = True
RNN_model.add(SimpleRNN(128,
                        activation='tanh',
                        return_sequences=False))
RNN_model.add(Dense(1, activation='sigmoid'))
```

```python
# printing/ model summery
print(RNN_model.summary())
RNN_model.compile(loss='binary_crossentropy',
                 optimizer='adam',
                 metrics=['accuracy'])
# Training the model
histroy = RNN_model.fit(x_train, y_train,
                       batch_size=64,
                       epochs=5,
                       validation_data=(x_valid, y_valid))
```

```python
# printing model score on test data
print()
print("Simple_RNN Score-->",RNN_model.evaluate(x_test,y_test,verbose=0))
```
