```python
# Encoding-> using  this method wecan  convert our categorical data into the numerical data.
# 1. LABEL encoding->> transform catagorical data into numeriacal data but it will work only single column .
#  it will autometically decide which value pass to the which sub-catagory
# 2. ordinal encoding->transform catagorical data into numeriacal data but it will work multiple columns.
# it will autometically decide which value pass to the which sub-catagory
# 3. Onehot encoding-> transform catagorical data into numeriacal data but it will create new column based on sub-catogorise which you mentioned in your column
# it will not provide you coulmn name.
# 4. get_dummies->  transform catagorical data into numeriacal data but it will create new column based on sub-catogorise which you mentioned in your column
```

```python
# 1..label encoding
import pandas as pd
import numpy as np
df=pd.read_csv('/content/covid_toy.csv')
df.head()
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer()
df['fever']=si.fit_transform(df[['fever']])
df.head()
```

```python
df['city'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['gender']=le.fit_transform(df['gender'])
df['cough']=le.fit_transform(df['cough'])
df['city']=le.fit_transform(df['city'])
df['has_covid']=le.fit_transform(df['has_covid'])
df.head()
df.sample(3)
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_train_new=pd.DataFrame(x_train_sc,columns= x_train.columns)
x_train_new.describe()
np.round(x_train_new.describe(), 2)
# df['city'].value_counts()
x_train_new.head()
```

```python
df=pd.read_csv("/content/tips.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
df['smoker']=le.fit_transform(df['smoker'])
df['day']=le.fit_transform(df['day'])
df['time']=le.fit_transform(df['time'])
```

```python
df.head()
```

```python
x=df.drop(columns=['total_bill'])
y=df['total_bill']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x_train_mn=sc.fit_transform(x_train)
x_train_new=pd.DataFrame(x_train_mn, columns= x_train.columns)
x_train_new.describe()
```

```python
#2 ordinal encoder
df=pd.read_csv("/content/covid_toy.csv")
df.head()
```

```python
df=df.drop(columns=['age','fever'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Male','Female'],
                              ['Mild','Strong'],
                              ['Kolkata','Bangalore','Delhi','Mumbai'],
                              ['Yes','No']])

oe_sc=oe.fit_transform(df)
```

```python
new_df=pd.DataFrame(oe_sc,columns=df.columns)
new_df.head(3)
```
