```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/Attrition (1).csv")
```

```python
df
```

```python
import tensorflow as tf ##deep learining library
from sklearn.model_selection import train_test_split ##data dividation
from sklearn.preprocessing import StandardScaler,LabelEncoder ##data preprocessing
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
df=df.dropna()
```

```python
x=df.drop(columns=["Attrition"])
y=df["Attrition"]
```

```python
x.info()
```

```python
cf=['BusinessTravel',
    'Department',
    'EducationField',
    'Gender',
    'JobRole',
    'MaritalStatus',
    'Over18',
    'OverTime',


    ]
```

```python
x_encoded=pd.get_dummies(x,columns=cf,drop_first=True)
```

```python
x_encoded=x_encoded.drop(columns=["EmployeeCount","EmployeeNumber","StandardHours"])
```

```python
lb=LabelEncoder()
```

```python
y_encoded=lb.fit_transform(y)
```

```python
y_encoded
```

```python
x_train,xtest,y_train,ytest=train_test_split(x_encoded,y_encoded,test_size=0.2,random_state=42)
```

```python
sc=StandardScaler()
```

```python
x_train_scaled=sc.fit_transform(x_train)
xtest_scaled=sc.transform(xtest)
```

```python
# architecture of model
model=tf.keras.models.Sequential(
    [
        tf.keras.layers.Dense(128,activation="relu",input_dim=(x_train_scaled.shape[1])),
        tf.keras.layers.Dropout(0.3),#input layer
        tf.keras.layers.Dense(64,activation="relu"),
        tf.keras.layers.Dropout(0.3),#hidden layer
        tf.keras.layers.Dense(1,activation="sigmoid"),
        #output layer
    ]
)
```

```python
# Now compile model
loss_fn=tf.keras.losses.BinaryCrossentropy()
metrics=["accuracy"]
```

```python
learning_rate=0.001
momentum=0.9
optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate,
                                  momentum=momentum,nesterov=True)
```

```python
model.compile(optimizer=optimizer,loss=loss_fn,metrics=metrics)
```

```python
# Train model

```

```python
model.fit(x_train_scaled,y_train,epochs=50,batch_size=8,validation_split=0.1)
```

```python
loss,accuracy=model.evaluate(xtest_scaled,ytest)
```

```python
print("Total Accuracy",accuracy)
```
